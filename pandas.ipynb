Pandas-DataFrame And Series:- Pandas is a powerful data manupulation library in python,widely used for data analysis and data cleaning.It provides two primary data structures: Series and DataFrame. A Series is a one-dimensional array-like object,while a DataFrame is a two-dimensional,size-mutable,and potentially heterogeneous tabular data structure with labeled axes(rows and columns).


```python
import pandas as pd
```

```python
##Series
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print("Series\n",series)
print(type(series))
```

```python
##basic plotting with seaborn
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter Plot')
plt.show()
```

```python
sns.lineplot(x='size',y='tip',data=tips)
plt.title("Line plot")
```

```python
##categorical plots
sns.barplot(x='day',y='total_bill',data=tips)
plt.title("Bar plot")
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.title("Box plot")
plt.show()
```

```python
##violin plot

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title("Violin plot")
plt.show()
```

```python
##histogram
sns.histplot(x='total_bill',data=tips)
plt.title("Histogram")
plt.show()
```

```python
##KDE PLOT
sns.kdeplot(x='total_bill',data=tips,fill = True)
plt.title("KDE Plot")
plt.show()
```

```python
#Pairplot
sns.pairplot(tips)
plt.show()
```

```python
## Heatmap
corr= tips[['total_bill','tip','size']].corr()
sns.heatmap(corr,annot=True)
plt.show()
```

SQL AND SQLite
SQL is a standard language for managing and manipulating relational databases.SQLite is a self-contained,serverless,and zero-configuration database engine that is widely used for embedded database systems.In this lesson,we will cover the basics of SQL and SQLite,including creating databases,tables,and performing various SQL operations.

```python
import sqlite3
```

```python
##connect to an SQLite database
connection=sqlite3.connect('example.db')
connection
```

```python
cursor = connection.cursor()
```

```python
##create a table
cursor.execute('''
    Create Table If Not Exists employees(
        id Integer Primary Key,
        name text Not Null,
        age Integer,
        department text
    )
''')

## Commit the changes
connection.commit()
```

```python
cursor.execute('''
    Select * from employees


''')
```

```python
## Insert the data in the sqlite table
cursor.execute('''
Insert Into employees(name,age,department)
Values('Annada',20,'Data Scientist')

''')
cursor.execute('''
Insert Into employees(name,age,department)
Values('Anna',19,'Engineer')

''')
cursor.execute('''
Insert Into employees(name,age,department)
Values('Annad',21,'App developer')

''')
connection.commit()
```

```python
##Query the data from the table
cursor.execute('''
Select * from employees
''')
rows = cursor.fetchall()

## print the quaried data
for row in rows:
    print(row)
```

```python
## Update the data in the table
cursor.execute('''
UPDATE employees
Set age = 100
where name="Anna"
''')
connection.commit()
```

```python
##Query the data from the table
cursor.execute('''
Select * from employees
''')
rows = cursor.fetchall()

## print the quaried data
for row in rows:
    print(row)
```

```python
from sqlite3.dbapi2 import connect
connection = sqlite3.connect('example.db')  ## Connect to the database
cursor = connection.cursor()  ## Create a cursor object

#create a table for sales data

cursor.execute('''
Create Table If Not Exists sales(
    id Integer Primary Key,
    product_name text Not Null,
    quantity Integer
)
''')
sales_data = [
    ('Product A',100),
    ('Product B',200),
    ('Product C',150)
]

cursor.executemany('''
Insert Into sales(product_name,quantity)
Values(?,?)
''',sales_data)
connection.commit()
```

```python
##Query the data from the table
cursor.execute('''
Select * from sales
''')
rows = cursor.fetchall()

## print the quaried data
for row in rows:
    print(row)
```

```python
## close the connection
connection.close()
```
